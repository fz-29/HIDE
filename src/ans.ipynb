```python
import numpy as np
```

```python
def get_angle(prog):
    anc = prog*360/100.0
    if anc<0:
        return anc+360.0
    return anc
```

```python
def compute_phi(x,y):
    val = np.degrees(np.arctan2(x,y))
    if val<0:
        val+=360
    return val
```

```python
f = open('progress_pie.txt','r')
f2 = open('output.txt','w+')
num_cases = f.readline()
for ix in range(int(num_cases)):
    k = f.readline()
    arr = k.split()
    prog = int(arr[0])
    x = int(arr[1])
    y = int(arr[2])
    ang = get_angle(prog)
    
    R = 50.0 + 0.000001
    x-=50
    y-=50
    r = np.sqrt(x**2 + y**2)
    arcang = compute_phi(x,y)
#print arcang
    if arcang<=ang and r<=R:
        ans = "Case #"+str(ix+1)+": black"
    else:
        ans = "Case #"+str(ix+1)+": white"
    f2.write(ans)
    f2.write("\n")
f.close()
f2.close()
```
